# create a dataset

```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

# Set random seed for reproducibility
np.random.seed(42)

# Generate 5000 transactions
n_rows = 5000

# Date range: Jan 2023 to Apr 2024
start_date = datetime(2023, 1, 1)
dates = [start_date + timedelta(days=np.random.randint(0, 485)) for _ in range(n_rows)]

# Create dataset
data = {
    'transaction_id': range(1, n_rows + 1),
    'order_date': dates,
    'customer_id': np.random.randint(1001, 1501, n_rows),
    'product_id': np.random.randint(1, 101, n_rows),
    'product_name': np.random.choice(['Laptop', 'Phone', 'Tablet', 'Headphones', 'Charger', 
                                       'Mouse', 'Keyboard', 'Monitor', 'Webcam', 'Speaker'], n_rows),
    'category': np.random.choice(['Electronics', 'Accessories', 'Computing'], n_rows),
    'quantity': np.random.randint(1, 11, n_rows),
    'unit_price': np.random.uniform(10, 1500, n_rows).round(2),
    'discount_pct': np.random.choice([0, 5, 10, 15, 20, 25], n_rows),
    'shipping_cost': np.random.choice([0, 5.99, 9.99, 15.99], n_rows),
    'payment_method': np.random.choice(['Credit Card', 'Debit Card', 'PayPal', 'Cash'], n_rows),
    'region': np.random.choice(['North', 'South', 'East', 'West'], n_rows),
    'store_id': np.random.randint(101, 111, n_rows),
    'is_returned': np.random.choice([0, 1], n_rows, p=[0.92, 0.08]),
    'customer_rating': np.random.choice([1, 2, 3, 4, 5, np.nan], n_rows, p=[0.02, 0.03, 0.10, 0.35, 0.45, 0.05])
}

df = pd.DataFrame(data)

# Add some intentional missing values
df.loc[np.random.choice(df.index, 100, replace=False), 'unit_price'] = np.nan
df.loc[np.random.choice(df.index, 80, replace=False), 'discount_pct'] = np.nan

# Save to CSV
df.to_csv('ecommerce_sales.csv', index=False)

print("Dataset created successfully!")
print(f"Shape: {df.shape}")
print(df.head())
```

```python
# Fill missing unit_price with median
df['unit_price'].fillna({'unit_price':df['unit_price'].median(),'inplace':True})
```

```python
# Fill missing discount_pct with 0
df['discount_pct'].fillna({'discount_pct':0,'inplace':True})
```

```python
# Fill missing customer_rating with mode
df['customer_rating'] = df['customer_rating'].fillna(df['customer_rating'].mode()[0])
```

```python
df.isnull().sum()
```

```python
# Q3: Basic Calculations
# Create new columns:

# revenue = quantity × unit_price
df['revenue']=df['quantity']*df['unit_price']
```

```python
# discount_amount = revenue × (discount_pct / 100)
df['discount_amount']=df['revenue']*(df['discount_pct']/100)
```

```python
# net_revenue = revenue - discount_amount
df['net_revenue']=df['revenue']-df['discount_amount']
```

```python
# total_cost = net_revenue + shipping_cost
df['total_cost']=df['net_revenue']+df['shipping_cost']
```

```python
# Q4: Filtering
# Filter the data to show:

# All orders where revenue > $500
df[df['revenue']>500]
# Orders from 'North' region with quantity > 5
df[(df['region']=='North') & (df['quantity']>5)]
# Transactions paid via 'Credit Card' OR 'PayPal'
df[(df['payment_method']=='Credit Card') | (df['payment_method']=='PayPal')]
# Orders that were NOT returned
df[df['is_returned']==0]
```

```python
# Q5: Sorting

# Sort by net_revenue in descending order
df['net_revenue'].sort_values(ascending=False)
# Show top 20 highest revenue transactions
df['net_revenue'].nlargest(10)
# Sort by order_date (oldest first), then by net_revenue (highest first)
df['order_date']=pd.to_datetime(df['order_date'])
df.sort_values(by=['order_date','net_revenue'],ascending=[True,False]).tail(5)
```
