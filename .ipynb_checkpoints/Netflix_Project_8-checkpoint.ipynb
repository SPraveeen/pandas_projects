## 2. In which year highest no.of tv shows & movies were released ? show with Bar Graph

```python
df.dtypes
```

```python
# change release year colum to date time
df['Date_N']=pd.to_datetime(df['Release_Date'])
```

```python
# error removing
df['Date_N']=pd.to_datetime(df['Release_Date'].str.strip())
```

```python
# counts by dt.year.value_counts()
df['Date_N'].dt.year.value_counts()
```

```python
# bar chart
df['Date_N'].dt.year.value_counts().plot(kind='bar')
```
